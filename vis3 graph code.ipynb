```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import datetime as dt
```

```python
rent_df = pd.read_csv('rentprice.csv')
data_df = pd.read_csv('boston_rent.csv')
```

```python
hi = rent_df.plot.bar(x = 'Neighborhood', \
                      y = 'Average_Rent', color = 'lightblue', \
                          title = 'Average Rent Cost per neighborhood')
plt.show()
```

```python
data_df = data_df.dropna()
data_df['date'] = pd.to_datetime(data_df['date'])
data_df = data_df[data_df["date"].dt.year == 2022]
data_df = data_df.replace('Boston', 'BackBay')
data_df.info()
```

```python
data = data_df[['neighborhood', 'violation_type', 'latitude', 'longitude']]

#new_data = data.loc[data['neighborhood'].isin(['BackBay', 'Jamaica Plain', 'Hyde Park'])]

hello = data['neighborhood'].value_counts().plot(kind = 'bar', color = 'lightblue',
                                                     xlabel = 'Neighborhoods', \
                                                         ylabel = 'number of violations',\
                                                             title = 'Number of Violations in 2022')

plt.show()
```

```python
hey = data['violation_type'].value_counts().plot(kind='bar', color = 'lightblue',\
                                                 xlabel = 'Violations', \
                                                     ylabel = 'number of violations',\
                                                         title = 'Total number of each violations')
plt.show()

#jamaica_df = data[data["neighborhood"] == 'Jamaica Plain']

#hyde_df = data[data["neighborhood"] == 'Hyde Park']
```
